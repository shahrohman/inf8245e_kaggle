```python
import pandas as pd

train_data = pd.read_csv('train_cleaned2.csv')
train_data2 = pd.read_csv('train_cleaned.csv')
old_train_data = pd.read_csv('train.csv')
```

```python
# Compare different rows between train_data and train_data2 and print the rows that are different
count = 0
for i in range(len(train_data)):
    if train_data.iloc[i].equals(train_data2.iloc[i]) == False:
        print(i)
        print(train_data.iloc[i])
        print(train_data2.iloc[i])
        print(old_train_data.iloc[i])
        print('')
        count += 1
        
print(count)
```
